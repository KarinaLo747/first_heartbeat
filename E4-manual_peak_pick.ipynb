```python
import glob
import pathlib
import pandas
import pandas as pd
import first_heartbeat.circles_analysis as circle
from first_heartbeat.constants import manual_peak_find_csv
```

```python
embryo: int = 4
```

```python
with open('E' + str(embryo) + '-' + manual_peak_find_csv, 'w') as file:
    file.write('dir_path\n')

csv_files: list[str] = glob.glob(f'data/raw/*/E{embryo}/*/*.csv')

lst_of_results_dict: list[dict[str, any]] = []

for csv in csv_files:
    data_dir: str = str(pathlib.Path(csv).parent) + '/'
    results_dict: dict[str, any] = circle.manual_peak_pick(data_dir=data_dir, filter_regex='Mean')
    lst_of_results_dict.append(results_dict)

results_df: pandas.DataFrame = pd.DataFrame(lst_of_results_dict)
results_df.to_csv(f'E{embryo}-results.csv', index=False)

print('---> Finished <---')
```
