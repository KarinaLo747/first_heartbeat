# Direction analysis

```python
import pandas as pd
import glob
```

## Load data and concatonate

```python
csv_dir = 'data/processed/dataset/by_embryo/'
csv_files = glob.glob(csv_dir + '*results.csv')

df_lst = []

for filepath in csv_files:
    df = pd.read_csv(filepath)
    df_lst.append(df)

full_dataset = pd.concat(df_lst, ignore_index=True)

full_dataset.to_csv('data/processed/dataset/full_dataset.csv')

df = pd.read_csv('data/processed/dataset/full_dataset.csv', index_col=0)

df
```
